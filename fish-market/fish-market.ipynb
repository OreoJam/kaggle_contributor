### Preprocessing

```python
# seperate input and target
target_data = df['Species']
input_data = df.drop('Species', axis=1)
```

```python
target_data, input_data
```

```python
# encording
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
target_data = encoder.fit_transform(target_data)
target_map = {index:label for index, label in enumerate(encoder.classes_)}
target_map
```

```python
# polynomial features
'''
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(include_bias=False)
poly.fit(input_data)
poly_input_data = poly.transform(input_data)
poly_input_data.shape
'''
```

```python
# standardize
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(input_data)
scaled_input_data = ss.transform(input_data)
scaled_input_data[:10]
```

### Training & Testing

```python
# split train / test data
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(scaled_input_data, target_data, random_state=0)
```

```python
test_input.shape, train_input.shape
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_input, train_target)
print(lr.score(train_input, train_target), lr.score(test_input, test_target))
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
lgr.fit(train_input, train_target)
print(lgr.score(train_input, train_target), lgr.score(test_input, test_target))
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=13, random_state=0)
dt.fit(train_input, train_target)
print(dt.score(train_input, train_target), dt.score(test_input, test_target))
```

```python
# graph - decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 15))
plot_tree(dt, filled=True, feature_names=['Weight', 'Height', 'LenVertical', 'LenDiagonal', 'LenCross', 'Width'])
plt.show()
```
